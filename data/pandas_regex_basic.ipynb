````python
import pandas as pd

# pd.set_option('display.max_rows', 50)
import ipywidgets as widgets
from ipywidgets import interact

import warnings

warnings.filterwarnings("ignore")
````

````python
print(f"pandas version: {pd.__version__}")

pd.Timestamp.now()
````

````python
url = "https://github.com/prasertcbs/basic-dataset/raw/master/IMDBtop1000.csv"
df = pd.read_csv(url, usecols=["Title", "Actors"])
df
````

## interactive playground


````python
@interact
def filter(regex_pat=r"."):
    return df[
        (df["Title"].str.contains(regex_pat, regex=True, case=False))
    ].sort_values("Title", ascending=True)
````

## basic regex


````python
df[df["Title"].str.contains("man")]  # case sensitive
````

````python
df[df["Title"].str.contains("Man")]
````

````python
df[df["Title"].str.contains(r"man", case=False)]
````

## starts with


````python
df[df["Title"].str.contains(r"^man", case=False)]
````

## ends with


````python
df[df["Title"].str.contains(r"man$", case=False)]
````

## word boundary


````python
df[df["Title"].str.contains(r"man", case=False)]
````

## raw string


````python
df[df["Title"].str.contains("man\b", case=False)]  # case insenstive comparison
````

````python
df[df["Title"].str.contains("man\\b", case=False)]  # case insenstive comparison
````

````python
df[df["Title"].str.contains(r"man\b", case=False)]  # case insenstive comparison
````

````python
df[df["Title"].str.contains(r"\bman$", case=False)]  # r' (raw string)
````

````python
df[df["Title"].str.contains(r"^man", case=False)]
````

````python
df[df["Title"].str.contains(r"^man\b", case=False)]
````

````python
df[df["Title"].str.contains(r"man\b", case=False)]  # man + punctuation marks
````

````python
df[df["Title"].str.contains(r"man\b$", case=False)]
````

````python
df[df["Title"].str.contains(r"\bman\b", case=False)]
````

## match one of a set of characters [aeiou]


````python
# df[df['Title'].str.contains(r'[1234you]', case=False)]
df[df["Title"].str.contains(r"[1-4you]", case=False)]
````

````python
df[df["Title"].str.contains(r"wom[ae]n$", case=False)]
````

## OR


````python
df[df["Title"].str.contains(r"bat|spider|ant|iron", case=False)]
````

````python
df[df["Title"].str.contains(r"(bat|spider|ant|iron)man", case=False)]
````

````python
df[df["Title"].str.contains(r"(bat|spider|ant|iron)\s*man", case=False)]
````

````python
df[df["Title"].str.contains(r"(bat|spider|ant|iron)[\-\s]*man", case=False)]
````

## group


````python
df[df["Title"].str.contains(r"^bat|^spider|^iron|^ant", case=False)]
````

````python
df[df["Title"].str.contains(r"^(bat|spider|iron|ant)", case=False)]
````

## \d for digit


````python
df[df["Title"].str.contains(r"\d", case=False)]
````

````python
df[df["Title"].str.contains(r"[0-9]", case=False)]
````

````python
df[df["Title"].str.contains(r"\d{2}", case=False)]
````

````python
df[df["Title"].str.contains(r"\d{2}$", case=False)]
````

````python
df[df["Title"].str.contains(r"(3|iii|three)$", case=False)]
````

````python
df[df["Title"].str.contains(r"\b(3|iii|three)$", case=False)]
````

````python
df[
    (df["Title"].str.contains("iron|avenger", case=False))
    & (df["Actors"].str.contains("chris", case=False))
]
````

## AND with lookahead ?=


````python
df[df["Actors"].str.contains(".*Downey.*Evans", case=False)]
````

````python
df[df["Actors"].str.contains(".*Evans.*Downey", case=False)]
````

````python
df[df["Actors"].str.contains("(.*Evans.*Downey)|(.*Downey.*Evans)", case=False)]
````

````python
df[df["Actors"].str.contains("(?=.*Downey)(?=.*Evans)", case=False)]
````

````python
df[df["Actors"].str.contains("(?=.*Evans)(?=.*Downey)", case=False)]
````

## quantifiers

```
- *, ?, +
- a* -> a{0,}
- a+ -> a{1,}
- a? -> a{0,1}
```


````python
df[df["Title"].str.contains(r"i*", case=False)]
````

````python
df[df["Title"].str.contains(r"i{0,}", case=False)]
````

````python
df[df["Title"].str.contains(r"i?", case=False)]
````

````python
df[df["Title"].str.contains(r"i{0,1}", case=False)]
````

````python
df[df["Title"].str.contains(r"i+", case=False)]
````

````python
df[df["Title"].str.contains(r"i{1,}", case=False)]
````

````python
df[df["Title"].str.contains(r"i{2,}", case=False)]
````

````python
df[df["Title"].str.contains(r"\bi{2,}", case=False)]
````

````python
df[df["Title"].str.contains(r"\bi{2}\b", case=False)]
````
