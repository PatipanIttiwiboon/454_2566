```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format='retina'

import warnings
warnings.filterwarnings('ignore')
```

ref: https://en.wikipedia.org/wiki/Rose_(mathematics)

$$
x=\cos(k\theta)\cos(\theta)\\
y=\cos(k\theta)\sin(\theta)
$$


```python
k = 10
j = 3
theta = np.linspace(0, 2 * np.pi, 720)
x = np.cos(k * theta) * np.cos(j * theta)
y = np.cos(k * theta) * np.sin(j * theta)
plt.figure(figsize=(7, 7))
plt.plot(x, y)
```

```python
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
```

```python
@interact(
    k=widgets.IntSlider(min=2, max=30, step=1, value=6),
    j=widgets.IntSlider(min=2, max=30, step=1, value=1),
)
def plot_flower(k, j):
    theta = np.linspace(0, 2 * np.pi, 720)
    x = np.cos(k * theta) * np.cos(j * theta)
    y = np.cos(k * theta) * np.sin(j * theta)
    plt.figure(figsize=(7, 7))
    plt.plot(x, y)
```

```python
@interact(
    k=widgets.IntSlider(min=2, max=30, step=1, value=6),
    j=widgets.IntSlider(min=2, max=30, step=1, value=1),
    p=widgets.FloatSlider(min=0.1, max=2, step=0.1, value=1),
)
def plot_flower(k, j, p):
    n_step = 15
    plt.figure(figsize=(7, 7))
    for r in range(0, 360, n_step):
        theta = np.radians(np.linspace(r, r + n_step, 180))
        x = np.cos(k * theta) * np.cos(j * theta) ** p
        y = np.cos(k * theta) * np.sin(j * theta) ** p
        plt.plot(x, y)
```
